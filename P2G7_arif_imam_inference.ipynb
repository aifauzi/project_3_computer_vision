```python
# import library

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
```

```python
model_cnn = load_model('/content/drive/MyDrive/Model/model_cnn_tl.keras')
```

```python
# buat function untuk memprediksi gambar

def prediction(file):
  img = tf.keras.utils.load_img(file, target_size=(224, 224))
  #x = tf.keras.utils.img_to_array(img)/255

  plt.imshow(img)

  x = np.expand_dims(img, axis=0)

  images = np.vstack([x])
  classes = model_cnn.predict(images, batch_size=128)
  idx = np.argmax(classes)
  clas = ['bacterialblight', 'brownspot', 'leafsmut']
  print('Prediction is a {}'.format(clas[idx]))
```
